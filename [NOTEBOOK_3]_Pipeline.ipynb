## Criação de ETL e métodos organizados para monitoramento e automação

```python
# Configuração do logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
```

```python
# Cabeçalhos da API
headers = {"chave-api-dados": "2de52c74c78e0ab8482ae9b3008e71f2"}
```

```python
# Função para fazer a requisição e retornar a resposta JSON
def run_request(url_):
    resposta = requests.get(url_, headers=headers)
    return resposta.json()
```

```python
# Função para coletar dados a partir das URLs
def coletar_dados(url):
    logging.info('Pegou URL para a criação das diferentes URLs criando uma para cada página')
    list_urls = [url + str(i) for i in range(1, 6)]  # Exemplo limitado a 5 páginas
    arr_urls = np.asarray(list_urls)

    vec_run = np.vectorize(run_request)
    logging.info('Fazendo uma requisição para cada URL')
    arr_responses = vec_run(arr_urls)
    logging.info('Requisições concluídas')

    # Convertendo as respostas para uma lista
    arr_responses = arr_responses.tolist()

    # Concatenando todas as respostas
    resposta = np.concatenate(arr_responses).tolist()
    return resposta
```

```python
# Função para transformar os dados em um DataFrame e calcular proporções
def transformar_dado(dado):
    logging.info('Criando dataframe')

    # Cria o DataFrame a partir do dado
    df = pd.DataFrame(dado)

    logging.info('Selecionando colunas relevantes')
    df_area = df[['funcao', 'valorEmpenhado']].copy()  # Use .copy() para evitar o "SettingWithCopyWarning"

    logging.info('Transformando valores de string para float')
    df_area.loc[:, 'valorEmpenhado'] = df_area['valorEmpenhado'].str.replace('.', '').str.replace(',', '.').astype(np.float64)

    logging.info('Transformando dado string -> float')
    agg_area = df_area.groupby('funcao').sum('valorEmpenhado')

    logging.info('Transformando dados para retirar proporção em porcentagem')
    agg_area = agg_area['valorEmpenhado'] / df_area['valorEmpenhado'].sum()

    logging.info('Resetando o índice do DataFrame agregado')
    agg_area = agg_area.reset_index()

    return agg_area
```

```python
# Função para salvar os dados em um arquivo CSV
def carga(dado):
    # Converte os dados para um DataFrame, caso não seja um
    if isinstance(dado, list):
        dado_df = pd.DataFrame(dado)
    elif isinstance(dado, pd.DataFrame):
        dado_df = dado
    else:
        raise ValueError("O parâmetro 'dado' deve ser uma lista de dicionários ou um DataFrame.")

    # Salva o DataFrame em um arquivo CSV
    dado_df.to_csv('distribuicao_empenho_area_20220.csv', index=False, encoding='utf-8')
    logging.info('Dados salvos em distribuicao_empenho_area_20220.csv')
```

**2. Fazer a extração dos dados**

Construir URLs ou consultas: montar URLs ou consultar SQL para especificar as informações que deseja extrair, geralmente de acordo com os parâmetros definidos.

Requisições ou consultas: executar as chamadas à API, banco de dados ou sistema de arquivos, processando uma ou múltiplas páginas.

Armazenamento temporário: armazenar as respostas em uma estrutura temporária (listas ou arrays) para facilitar o processamento posterior.

**3. Transformar e limpar os dados**

Estruturar os dados: converter as respostas para uma estrutura tabular (como um DataFrame) para manipulação mais eficiente.

Agregação e cálculo de métricas: realizar agregações (soma, média, proporções) e outras transformações conforme o objetivo da análise.

**4. Organizar e padronizar os dados**

**5. Armazenar os dados processados**

Salvar em arquivo: exportar os dados processados para um formato final (como CSV ou Excel) para garantir que possam ser usados ou analisados posteriormente.

Verificação de integridade: antes de encerrar, verificar a integridade dos dados, como o número de registros e se as colunas estão completas e corretas.
