2. Qual é a média, a moda e a mediana de horas de sono para cada uma
das profissões? [‘mean’, np.median, pd.Series.mod]

```python
df_media_sono = df_saude.groupby('Profissão')['Duração do sono'].agg([
    ('Média', 'mean'),
    ('Mediana', 'median'),
    ('Moda', pd.Series.mode)
])

print(df_media_sono)
```

3. Das pessoas que atuam com engenharia de software qual a
porcentagem de obesos?

```python
# Filtrando pela profissão
eng_software = df_saude[df_saude['Profissão'] == 'Eng. de Software']

# Carregando número total de participantes dessa profissão
num_total_engenheiros = eng_software.shape[0]

# Número de engenheiros obesos
num_obesos = eng_software[eng_software['Categoria IMC'] == 'Obesidade'].shape[0]

# Calculando a porcentagem
porcentagem_obesos = (eng_software['Categoria IMC'] == 'Obesidade').mean() * 100

print(f'Número total de engenheiros de software: {num_total_engenheiros}')
print(f'Número de engenheiros de software obesos: {num_obesos}')
print(f'Porcentagem de engenheiros de software obesos: {porcentagem_obesos}%')
```

4. De acordo com os dados, advogar ou ser representante de vendas faz
você dormir menos? (Use o método ‘isin’, considere a média)


```python
# Calculando média da amostra
media_sono_todos = df_media_sono['Média'].mean()

# Puxar a média dos dois da tabela feita no exercício anterior
media_advogado = df_media_sono.loc['Advogado(a)', 'Média']
media_vendas = df_media_sono.loc['Representante de Vendas', 'Média']

print(f'A média de sono da amostra é de {media_sono_todos:.2f} horas')
print(f'A média de sono dos advogados é de {media_advogado:.2f} horas')
print(f'A média de sono dos representantes de vendas é de {media_vendas:.2f} horas')

if media_advogado < media_sono_todos:
    print("Advogados dormem menos que a média da amostra.")
else:
    print("Advogados não dormem menos que a média da amostra.")

if media_vendas < media_sono_todos:
    print("Representantes de vendas dormem menos que a média da amostra.")
else:
    print("Representantes de vendas não dormem menos que a média da amostra.")
```

5. Entre quem fez enfermagem e quem fez medicina, quem tem menos
horas de sono? (Use o método ‘isin’, considere a média)


```python
# Filtrar enfermeiros e médicos
enf_med = df_saude[df_saude['Profissão'].isin(['Enfermeiro(a)', 'Médico(a)'])]

# Calcular a média de horas de sono
media_sono_enf_med = enf_med.groupby('Profissão')['Duração do sono'].mean()

# Puxar a média dos dois
media_enf = media_sono_enf_med.loc['Enfermeiro(a)']
media_med = media_sono_enf_med.loc['Médico(a)']

media_sono_enf_med

print(f'Enfermeiros(as) têm, em média, {media_enf:.2f} horas de sono por dia.')
print(f'Médicos(as) têm, em média, {media_med:.2f} horas de sono por dia.')

if media_enf > media_med:
    print("Vemos assim que médicos(as) têm, em média, menos horas de sono por dia.")
else:
    print("Vemos assim que enfermeiros(as) têm, em média, menos horas de sono por dia.")
```

8. Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)


```python
# Vi que é possível fazer um cálculo para obter um valor médio de pressão
# sanguínea, chamada Pressão Arterial Média (PAM = (sístole + 2 * diástole)/3)

# Dividir a coluna 'Pressão sanguínea' em 'Pressão Sistólica' e 'Pressão Diastólica'
df_saude[['Pressão Sistólica', 'Pressão Diastólica']] = df_saude['Pressão sanguínea'].str.split('/', expand=True)

# As colunas ainda estão como string, então precisa converter para numérico
df_saude['Pressão Sistólica'] = pd.to_numeric(df_saude['Pressão Sistólica'])
df_saude['Pressão Diastólica'] = pd.to_numeric(df_saude['Pressão Diastólica'])

# Calcular PAM
df_saude['PAM'] = (df_saude['Pressão Sistólica'] + 2 * df_saude['Pressão Diastólica']) / 3

# Calcular média da PAM por gênero
media_pam_genero = df_saude.groupby('Gênero')['PAM'].mean()

print("Média de Pressão Arterial Média (PAM):")
print(media_pam_genero)

if media_pam_genero['Homem'] > media_pam_genero['Mulher']:
    print("Homens têm maior Pressão Arterial Média.")
else:
    print("Mulheres têm maior Pressão Arterial Média.")
```

9. É predominante entre os participantes dormir 8 horas por dia
(considere usar Moda como medida)?


```python
# Calculando a moda da amostra
moda_horas_sono = df_saude['Duração do sono'].mode()[0]

print(f'A moda de sono da amostra é de {moda_horas_sono:.2f} horas')

if moda_horas_sono > 8:
    print("É predominante entre os participantes dormir 8 horas por dia.")
else:
    print("Não é predominante entre os participantes dormir 8 horas por dia.")
```

10. Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
media_passos_acima_70 = df_saude[df_saude['Frequência cardíaca'] > 70]['Passos diários'].mean()
media_passos_abaixo_ou_igual_70 = df_saude[df_saude['Frequência cardíaca'] <= 70]['Passos diários'].mean()

(media_passos_acima_70, media_passos_abaixo_ou_igual_70)

print(f'Pessoas com frequência cardíaca acima de 70 dão, em média, {media_passos_acima_70:.2f} passos por dia')
print(f'Pessoas com frequência cardíaca abaixo de 70 dão, em média, {media_passos_abaixo_ou_igual_70:.2f} passos por dia')

if media_passos_acima_70 > media_passos_abaixo_ou_igual_70:
    print("Pessoas com frequência cardíaca acima de 70 dão, em média, mais passos por dia.")
else:
    print("Assim, vemos que pessoas com frequência cardíaca abaixo de 70 dão, em média, mais passos por dia.")
```
